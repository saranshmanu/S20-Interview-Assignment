```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
```

```python
df = pd.read_csv("ratings.csv")
df = df.drop(["userId"], axis = 1)
df = df.loc[:, (df != 0).any(axis=0)]
X_train, X_test = train_test_split(df, train_size=0.8)
```

```python
df.describe()
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from keras.models import Sequential
from keras.layers import Dense
```

```python
# model = tf.keras.Sequential([
#     layers.Dense(256, activation='sigmoid', input_shape=(5379,)),
#     layers.Dense(5379, activation='linear')
# ])
model = Sequential()
model.add(Dense(246, input_dim=5379, activation='sigmoid'))
model.add(Dense(5379, activation='linear'))
```

```python
model.compile(optimizer=tf.train.AdamOptimizer(),
              loss='mse',
              metrics=['mse'])
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
```

```python
model.fit(X_train, X_train, batch_size=32, validation_data=(X_test, X_test), epochs=100)
```

```python
# sample_user = X_test[14,:].reshape(1, 5379)
sample_user = np.array([1]*5379).reshape(1, 5379)
prediction = model.predict(sample_user)
```

```python
movies = pd.read_csv("ml-latest-small/movies.csv")
movies_Id = movies.movieId
movies_title = movies.title
id2title = dict(zip(movies_Id, movies_title))
```

```python
id2score = dict(zip(df.columns.values, prediction[0]))
ranking = sorted(id2score.items(), key = lambda kv:(kv[1], kv[0]), reverse=True)
movie_rankings_with_name = []
for id, rank in ranking:
    if id in df.columns.values:
        print(id2title[int(id)])
#         movie_rankings_with_name = [id2title[i] ]
```
